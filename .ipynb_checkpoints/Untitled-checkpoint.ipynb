```python
df = pd.read_csv("Metadata.csv", sep = ",")
df_meta = pd.read_csv("Metadata view features.csv", sep = ",")
df_engagement = pd.read_csv("Engagement Features.csv", sep = ",")
```

```python
# Obtain all column names
first = set(df.columns)
second = set(df_meta.columns)
third = set(df_engagement.columns)
second = second - first
third = (third - second) - first
final = list(first) + list(second) + list(third)


# Merge all dataset together
dftemp = pd.concat([df, df_meta[second], df_engagement[third]], axis = 1)
df = dftemp
```

```python
# Checking if there's any column with only few values. 
to_drop = []
for i, col in enumerate(df.columns):
    if df[col].value_counts().shape[0] == 1:
        print(col, ": Skipped")
        to_drop.append(col)
    else:
        # print(col)
        df[col].value_counts()
df = df.drop(columns = to_drop)
```
